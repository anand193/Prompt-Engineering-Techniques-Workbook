```python
!pip install nltk rouge.score bert-score --quiet
import nltk
nltk.download('punkt_tab')
from nltk.translate.bleu_score import sentence_bleu,SmoothingFunction
from rouge_score import rouge_scorer
from bert_score import score
```

```python
reference = ['The quick brown fox jumps over the lazy dog']
candidate = 'The brown fox quickly jumps over the lazy dog'
```

```python
reference[0]
```

```python
reference_token = (nltk.word_tokenize(reference[0]))
```

```python
candidate_token = [nltk.word_tokenize(candidate)]
```

```python
blue1 = sentence_bleu([reference_token], candidate_token[0], smoothing_function=SmoothingFunction().method1)
print(f"BLEU-1 score: {blue1}")
```

```python
scorer = rouge_scorer.RougeScorer(['rouge1','rougeL'],use_stemmer = True)
```

```python
rouge = scorer.score(reference[0],candidate)
print(rouge)
```

## BERT SCORE

```python
import bert_score
refs = [reference[0]]
cands = [candidate]
P, R, F1 = score(cands, refs, lang="en", verbose=True)
```

```python
print(P.mean())
print(R.mean())
print(F1.mean())
```
